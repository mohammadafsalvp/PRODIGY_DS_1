```python
import matplotlib.pyplot as plt

# Data
ages = [22, 25, 30, 21, 28, 35, 40, 32, 19, 27, 29, 31, 26, 33, 24, 23, 36, 39, 37, 34]

# Plotting
plt.figure(figsize=(8, 6))
plt.hist(ages, bins=5, edgecolor='black', alpha=0.7)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Ages in the Population')
plt.grid(True)
plt.show()
```
